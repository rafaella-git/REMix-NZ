```python
gdx: GDXEval = GDXEval("remixplot/input/kohleatlas/museko_WSK.gdx")
trans_flow_df: pd.DataFrame = gdx["transfer_flows_annual"].reset_index()
links_df: pd.DataFrame = gdx["map_linksModel"].reset_index().drop("element_text", axis=1)
graph_kohleatlas: nx.Graph = nx.Graph([start_end.split("__") for start_end in links_df["linksData"]])
kohleatlas_df: gpd.GeoDataFrame = gpd.read_file('remixplot/input/kohleatlas/Kohleatlas.geojson')
kohleatlas_mapping: dict = json.load(open("remixplot/input/kohleatlas/Kohleatlas_mapping.txt"))
```

```python
fig: go.Figure = plot_network(
    graph_kohleatlas,
    kohleatlas_df,
    kohleatlas_mapping,
    contours_bool=True,
    nodes_bool=True,
    connection_bool=False,
)
fig.show()
```
